We have missing values in Age, Height, Weight, and Medal columns. 
The missing values in Age, Height and Weight are proportionally small (to the whole data set)
so I am going to fill them with their respective averages. 
The null values in Medal column represent no medal - 'None'.

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Height'] = df['Height'].fillna(df['Height'].mean()) 
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
df['Medal'] =df['Medal'].fillna('None') 
```

Let's now create a dataset with unique names

```python
array = df_medals['Name'].unique()
```

```python
df_unique_names = pd.DataFrame(array, columns = ['UniqueName'])
```

```python
df_unique_names.head()
```

Extracting surnames form list of unique names

```python
surnames = []
for x in df_unique_names['UniqueName']:
    if ',' in x:
        x = x.split(',')[0]
    x_st = x.split(' ')[-1]
    surnames.append(x_st) 
```

```python
surnames
```

```python
seen = {}
dupes = []

for x in surnames:
    if x not in seen:
        seen[x] = 1
    else:
        if seen[x] == 1:
            dupes.append(x)
        seen[x] += 1
```

Finding the duplicates from list of surnames

```python
dupes
```

Adding 'Surname' column to our df_medals

```python
surnames_general = []
for x in df_medals['Name']:
    if ',' in x:
        x = x.split(',')[0]
    x_st = x.split(' ')[-1]
    surnames_general.append(x_st)     
```

```python
df_medals['Surname'] = surnames_general
```

Merging duplicate surnames with df_medals to have access to more info

```python
df_dupes = pd.DataFrame(dupes, columns = ['Surname'])
```

```python
df_medals_dupes_merge = df_dupes.merge(df_medals, on='Surname', how='left')
```

```python
df_surname_grouped = df_medals_dupes_merge[['Surname', 'Medal']].groupby('Surname', as_index = False).count()
```

<b>Here we have the 5 surnames of families that won the most medals.<b>
